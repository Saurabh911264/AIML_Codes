```python
import random
class Environment(object):
    def __init__(self):
        # instantiate locations and conditions
        # 0 indicates  Clean and 1 indicates Dirty
        self.locationCondition = {'A': '0', 'B': '0'}
        self.obstacle={'A':'0','B':'0'}
        # randomize conditions in locations A and B
        self.locationCondition['A'] = random.randint(0, 1)
        self.locationCondition['B'] = random.randint(0, 1)
        self.obstacle['A']=random.randint(0, 1)
        self.obstacle['B']=random.randint(0, 1)
        #Battery Life of the vacuum cleaner
        self.Battery=2500


class SimpleReflexVacuumAgent(Environment):
  def __init__(self, Environment):
    user_input='0'
    while user_input !='1':

      if Environment.Battery>0:
        print (Environment.locationCondition)
        print(Environment.obstacle)
        # Instantiate performance measurement
        Score = 0
        # place vacuum at random location
        vacuumLocation = random.randint(0, 1)
        # if vacuum at A
        if vacuumLocation == 0:
            print ("Vacuum is randomly placed at Location A")
            # and A is Dirty
            if Environment.locationCondition['A'] == 1:
                print ("Location A is Dirty. ")
                # suck and mark clean
                if Environment.obstacle['A']==1:
                  print("Can't clean obstacle is ahead")
                  Environment.Battery-=100
                else:
                  Environment.locationCondition['A'] = 0;
                  Environment.Battery-=250
                  Score += 1
                  print ("Location A has been Cleaned.")

                # if B is Dirty
                if Environment.locationCondition['B'] == 1:
                    print ("Location B is Dirty.")
                    # move to B
                    print ("Moving to Location B...")
                    Environment.Battery-=150
                    Score -= 1
                    # suck and mark clean
                    if Environment.obstacle['B']==1:
                      print("Can't clean obstacle is ahead")
                      Environment.Battery-=100
                    else:
                      Environment.locationCondition['B'] = 0;
                      Environment.Battery-=250
                      Score += 1
                      print ("Location B has been Cleaned.")

            else:
                # if B is Dirty
                if Environment.locationCondition['B'] == 1:
                    print ("Location B is Dirty.")
                    # move to B
                    Score -= 1
                    print ("Moving to Location B...")
                    Environment.Battery-=150
                    # suck and mark clean
                    if Environment.obstacle['B']==1:
                      print("Can't clean obstacle is ahead")
                      Environment.Battery-=100
                    else:
                      Environment.locationCondition['B'] = 0;
                      Score += 1
                      print ("Location B has been Cleaned.")
                      Environment.Battery-=250

        elif vacuumLocation == 1:
            print ("Vacuum is randomly placed at Location B. ")
            # and B is Dirty
            if Environment.locationCondition['B'] == 1:
                print ("Location B is Dirty")
                # suck and mark clean
                if Environment.obstacle['B']==1:
                    print("Can't clean obstacle is ahead")
                    Environment.Battery-=100
                else:
                    Environment.locationCondition['B'] = 0;
                    Environment.Battery-=250
                    Score += 1
                    print ("Location B has been Cleaned.")

                # if A is Dirty
                if Environment.locationCondition['A'] == 1:
                    print ("Location A is Dirty")
                    # move to A
                    Score -= 1
                    print ("Moving to Location A")
                    Environment.Battery-=150
                    # suck and mark clean
                    if Environment.obstacle['A']==1:
                        print("Can't clean obstacle is ahead")
                        Environment.Battery-=100
                    else:
                        Environment.locationCondition['A'] = 0;
                        Environment.Battery-=250
                        Score += 1
                        print ("Location B has been Cleaned.")
            else:

                # if A is Dirty
                if Environment.locationCondition['A'] == 1:
                    print( "Location A is Dirty")
                    # move to A
                    print ("Moving to Location A")
                    Environment.Battery-=150
                    Score -= 1
                    # suck and mark clean
                    if Environment.obstacle['A']==1:
                        print("Can't clean obstacle is ahead")
                        Environment.Battery-=100
                    else:
                        Environment.locationCondition['A'] = 0;
                        Environment.Battery-=250
                        Score += 1
                        print ("Location A has been Cleaned.")
        # done cleaning
        print (Environment.locationCondition)
        print ("Performance Measurement: " + str(Score))
        Environment.locationCondition['A'] = random.randint(0, 1)
        Environment.locationCondition['B'] = random.randint(0, 1)
        Environment.obstacle['A']=random.randint(0, 1)
        Environment.obstacle['B']=random.randint(0, 1)
        user_input=input("Press 1 to stop")
      else:
        user_input='1'
        print("Battery is dead!!")

theEnvironment = Environment()
theVacuum = SimpleReflexVacuumAgent(theEnvironment)
# Obstacle Detection,Battery Life,User Input
```
